```python
from sklearn.cluster import DBSCAN
from sklearn.cluster import SpectralClustering
from sklearn.datasets import load_iris
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import silhouette_score
```

```python
dataset = load_iris()
```

```python
X = dataset.data
```

```python
spectral = SpectralClustering(n_clusters=3)
```

```python
spectral_predictions = spectral.fit_predict(X)
```

```python
# DBSCAN NOW
dbscan = DBSCAN()
```

```python
dbscan_predictions = dbscan.fit_predict(X)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
# Getting silhouette score
score_spectral = silhouette_score(X, spectral_predictions)
score_dbscan = silhouette_score(X, dbscan_predictions)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12,6))
plot_spectral = ax[0].scatter(X[:, 0], X[:, 1], c=spectral_predictions, cmap='viridis')
plot_dbscan = ax[1].scatter(X[:, 0], X[:, 1], c=dbscan_predictions, cmap='viridis')
ax[0].set_title('Spectral Clustering')
fig.suptitle('Unsupervised Techniques', fontsize=16, y=1.02)
ax[1].set_title('DBSCAN Clustering')
fig.colorbar(plot_spectral, ax=ax[0])
fig.colorbar(plot_dbscan, ax=ax[1])
```
